```
import numpy as np
import matplotlib.pyplot as plt
```

```
center_1 = np.array([1,1])
center_2 = np.array([5,5])

data_1 = np.random.randn(200,2) + center_1
data_2 = np.random.randn(200,2) + center_2


data = np.concatenate((data_1, data_2))

plt.scatter(data[:,0], data[:,1], s=7)
```

```
from sklearn.cluster import KMeans
kmns = KMeans(n_clusters = 2)
kmns.fit(data)
```

```
kmns.labels_
```

```
kmns.labels_.shape
```

```
for i in range(400):
  c='r'
  if kmns.labels_[i]==0:
    c='b'
  plt.scatter(data[i][0],data[i][1],c=c,s=7)
plt.show()
```

```
center_1 = np.array([1,1])
center_2 = np.array([5,5])
center_3 = np.array([8,1])

# Generate random data and center it to the three centers
data_1 = np.random.randn(100, 2) + center_1
data_2 = np.random.randn(100,2) + center_2
data_3 = np.random.randn(100,2) + center_3

data = np.concatenate((data_1, data_2, data_3))

plt.scatter(data[:,0], data[:,1], s=7)
```

```
from sklearn.cluster import KMeans
kmns = KMeans(n_clusters = 3).fit(data)


for i in range(300):
  if kmns.labels_[i]==0:
    c='b'
  elif kmns.labels_[i]==1:
    c='y'
  else:
    c='r'
  plt.scatter(data[i][0],data[i][1],c=c,s=7)
plt.show()
```
