```
import math
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
import matplotlib.pyplot as plt

from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```
url = "http://www.tinyurl.com/irisflowerdata"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv(url, names=names)
```

```
dataset.head()
```

```
data = dataset.values

train_data = data[0:35]
train_data = np.vstack([train_data,data[50:85]])
train_data = np.vstack([train_data,data[100:135]])

test_data = data[35:50]
test_data = np.vstack([test_data,data[85:100]])
test_data = np.vstack([test_data,data[135:150]])


print(train_data.shape, test_data.shape)
```

```
train_data = shuffle(train_data)
```

```
X_train = train_data[:,0:4]
Y_train = train_data[:,4]

X_test = test_data[:,0:4]
Y_test = test_data[:,4]


print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)

predictions = knn.predict(X_test)

print(accuracy_score(Y_test, predictions))
```

```
def KNN(lis):
  
  pred = None
  dist = 1000
  
  for i in range(len(X_train)):
    
    x = X_train[i] 
    temp_dist = math.sqrt((x[0]-lis[0])**2 + (x[1]-lis[1])**2 + (x[2]-lis[2])**2 + (x[3]-lis[3])**2)
    
    if temp_dist < dist:
      dist = temp_dist
      pred = Y_train[i]
      
  return pred
```

```
def predictions(X_test):
  results = []
  
  for x in X_test:
    results.append(KNN(x))
    
  return results
    
```

```
results = predictions(X_test)
results
```

```
print(accuracy_score(Y_test, results))
```
